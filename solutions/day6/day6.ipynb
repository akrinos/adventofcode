```python
# vector of values is the current state of the nearby lanternfish, which needs to be amended
def dailystep(vector_vals):
    # decrement all vector values
    vector_vals = [curr-1 for curr in vector_vals]
    
    # see how many of them need to spawn a new lanternfish (they have become negative)
    need_to_spawn = [curr for curr in range(len(vector_vals)) if vector_vals[curr] < 0]
    
    # let's set their counters back
    vector_vals = [6 if curr < 0 else curr for curr in vector_vals]
    
    # this is how many lanternfish we should add
    vector_vals.extend([8] * len(need_to_spawn))
    
    # now let's return the fish
    return vector_vals

def initialize_dict(vector_vals):
    init_vals = [0] * len(list(range(0,9)))
    dict_init = dict(zip([str(curr) for curr in list(range(0,9))], [0] * len(list(range(0,9)))))
    for val in vector_vals:
        dict_init[str(val)] = dict_init[str(val)] + 1
        
    return dict_init

# more efficient implementation of daily step
def dailystep_efficient(val_dict):
    # this is how many we need to spawn
    need_to_spawn = val_dict[str(0)]

    # the rest we mancala.
    for curr in range(8):
        val_dict[str(curr)] = val_dict[str(curr + 1)]
    #preserve_vals = list(val_dict.values())
    #val_dict[str(7)] = preserve_vals[8]
    #val_dict[str(6)] = preserve_vals[7]
    #val_dict[str(5)] = preserve_vals[6]
    #val_dict[str(4)] = preserve_vals[5]
    #val_dict[str(3)] = preserve_vals[4]
    #val_dict[str(2)] = preserve_vals[3]
    #val_dict[str(1)] = preserve_vals[2]
    #val_dict[str(0)] = preserve_vals[1]
    
    # reset the ones we're spawning.
    val_dict[str(6)] = val_dict[str(6)] + need_to_spawn
    
    # we add the spawns to the 8 category
    val_dict[str(8)] = need_to_spawn
    print(val_dict)
    
    return val_dict
    

def runsim(numberofdays, inputvector):
    for day in list(range(numberofdays)):
        inputvector = dailystep(inputvector)
        
    return inputvector

def runsim_efficient(numberofdays, inputvector):
    dict_out = initialize_dict(inputvector)
    for day in list(range(numberofdays)):
        dict_out = dailystep_efficient(dict_out)
        
    print(dict_out, flush=True)
    return np.sum(list(dict_out.values()))
```
