```python
def get_winning_combos(board):
    n_rows = len(board.index)
    n_cols = len(board.columns)
    diagonal_list_forward = []
    all_winning = []
    for row_c in range(n_rows):
        set_numbers = set(board.iloc[row_c,:])
        diagonal_list_forward.append(board.iloc[row_c,row_c])
        all_winning.append(set_numbers)
    diagonal_list_rev = []
    # whoops no diagonals
    #all_winning.append(set(diagonal_list_forward))
    for col_c in range(n_cols):
        set_numbers = set(board.iloc[:,col_c])
        diagonal_list_rev.append(board.iloc[n_rows - col_c - 1,col_c])
        all_winning.append(set_numbers)
        
    #all_winning.append(set(diagonal_list_rev))
    return all_winning

def mark_numbers(called_number, winning_combos):
    [curr.remove(called_number) if called_number in curr else curr for curr in winning_combos]
    return winning_combos

def count_moves(board, called_numbers, min_number, ret_sum = False, last_moves = False):
    winning_combos = get_winning_combos(board)
    number_moves = 0
    for num_called in called_numbers:
        number_moves = number_moves + 1
        if not last_moves:
            if number_moves > min_number:
                return np.inf
        winning_combos = mark_numbers(num_called, winning_combos) 
        if set() in winning_combos:
            #return number_moves
            # just sum rows; cols are dups
            if not ret_sum:
                return number_moves
            else:
                total_out = sum([sum(list(curr)) for curr in winning_combos[0:5]])
                return total_out, num_called
            
        
    return np.inf
```
