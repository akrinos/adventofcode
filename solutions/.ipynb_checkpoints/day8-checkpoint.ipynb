```python
def parse_input(in_string):
    
    clock_digits = in_string.split(" | ")[0].split(" ")
    output_values = in_string.split(" | ")[1].split(" ")
    
    return clock_digits, output_values

def parse_input_top(in_vect):
    for curr in in_vect.split("\n"):
        parse_input(curr)
        
def count_unique(output_val):
    return sum([[len(curr) for curr in output_val].count(num) for num in [7,4,2,3]])

# we take these vals from the random output and arrange them where they need to be.
# if they are missing, they are a "none"
##      1 1 
##.    2.  3
##.    2.  3 
##.     4 4
##.    5.  6
##.    5.  6
##.     7 7
def decode_clock(three_digit, two_digit, four_digit, seven_digit, six_digits):
    clock_positions = dict() # dictionary where each of the 1-7 clock positions actually corresponds to at the
    # moment
    ## start with the smallest. this must be the 1 config. 
    # we use the 6 to find the order
    available_choices = list(two_digit)
    for curr in six_digits:
        # it's the 6 if it has c but not b.
        if len(set(available_choices) - set(curr)) == 1:
            three_digit_num = list(set(available_choices) - set(curr))[0]
            six_digit_num = list(set(available_choices) - set(three_digit_num))[0]
            break
    clock_positions["3"] = three_digit_num
    clock_positions["6"] = six_digit_num
    
    ## now the 3-digit.
    clock_positions["1"] = list(set(list(three_digit)) - set(list(two_digit)))[0]
    # the other two digits will be 3 & 6 and should be the same as the 1-dig
    
    ## now the 4-digit
    
    ## to determine which is two and which is 4, we need to find the zero (has 2 but not 4)
    available_choices = list(set(list(four_digit)) - set(list(two_digit)))
    for curr in six_digits:
        if len(set(available_choices) - set(curr)) == 1:
            four_digit_num = list(set(available_choices) - set(curr))[0]
            two_digit_num = list(set(available_choices) - set(four_digit_num))[0]
            break
    clock_positions["2"] = two_digit_num
    clock_positions["4"] = four_digit_num
    # the others are 3 & 6
    
    ## now the seven-digit. we need to use the 9  to figure out which is which.
    
    available_choices = list(set(list(seven_digit)) - set(list(four_digit) + list(three_digit)))
    for curr in six_digits:
        if len(set(available_choices) - set(curr)) == 1:
            five_digit_num = list(set(available_choices) - set(curr))[0]
            seven_digit_num = list(set(available_choices) - set(five_digit_num))[0]
            break
    clock_positions["5"] = five_digit_num
    clock_positions["7"] = seven_digit_num
    # the others are 1,2,3,4,6
    
    return clock_positions

def g_p(number,clock_position):
    return clock_position[str(number)]

def create_clock_dict(clock_position):
    decoder = dict()
    reversed_decoder = dict()
    
    decoder["1"] = set([clock_position["3"], clock_position["6"]])
    # 1,3,4,5,7
    decoder["2"] = set([clock_position["1"],clock_position["3"],clock_position["4"],clock_position["5"],
                        clock_position["7"]])
    # 1,3,4,6,7
    decoder["3"] = set([clock_position["1"],clock_position["3"],clock_position["4"],clock_position["6"],
                        clock_position["7"]])
    # 2,3,4,6
    decoder["4"] = set([clock_position["2"],clock_position["3"],clock_position["4"],clock_position["6"]])
    
    # 1,2,4,6,7
    decoder["5"] = set([clock_position["1"],clock_position["2"],clock_position["4"],clock_position["6"],
                        clock_position["7"]])
    # 1,2,4,5,6,7
    decoder["6"] = set([clock_position["1"],clock_position["2"],clock_position["4"],clock_position["5"],
                        clock_position["6"],
                        clock_position["7"]])
    # 1,3,6
    decoder["7"] = set([clock_position["1"],clock_position["3"],clock_position["6"]])
    # 1,2,3,4,5,6,7
    decoder["8"] = set([clock_position["1"],clock_position["2"],clock_position["3"],
                        clock_position["4"],clock_position["5"],clock_position["6"],
                        clock_position["7"]])
    # 1,2,3,4,6,7
    decoder["9"] = set([clock_position["1"],clock_position["2"],clock_position["3"],
                        clock_position["4"],clock_position["6"],
                        clock_position["7"]])
    # 1,2,3,5,6,7
    decoder["0"] = set([clock_position["1"],clock_position["2"],clock_position["3"],
                        clock_position["5"],clock_position["6"],
                        clock_position["7"]])
    
    for curr in list(range(10)):
        reversed_decoder["".join(sorted(list(decoder[str(curr)])))] = str(curr)
    return decoder,reversed_decoder


def all_decode(final_output, clock_digits):
    two_digits = [curr for curr in clock_digits if len(curr) == 2]
    three_digits = [curr for curr in clock_digits if len(curr) == 3]
    four_digits = [curr for curr in clock_digits if len(curr) == 4]
    seven_digits = [curr for curr in clock_digits if len(curr) == 7]
    six_digits = [curr for curr in clock_digits if len(curr) == 6]
    clock_position = decode_clock(three_digits[0], two_digits[0], four_digits[0], seven_digits[0], six_digits)
    decoder,reversed_decoder = create_clock_dict(clock_position)
    final_string = ""
    for out in final_output:
        final_string = final_string + reversed_decoder["".join(sorted(list(set(out))))]
    return int(final_string)
```
